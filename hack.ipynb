```python
import pandas as pd
import random
import requests
from PIL import Image
from io import BytesIO
from sentence_transformers import SentenceTransformer
from PIL import Image
```

```python
data = pd.read_csv("nft.csv")
# data[data["Logo"].isnull()]
data.dropna(subset = "Logo")
data
```

```python
data = data[["Name", "Index", "Logo", "Volume", "Average_Price_USD"]]

categories = ["category_a", "category_b", "category_c"]
total_categories = len(categories)
for i in categories:
    data[i] = 0

data.head()
```

```python

# assigning random categories to database
for i in range(len(data)):
    cat_count = random.randint(0, total_categories) # number of categories that are to be allocated per row
    choosed_cat = random.sample(range(0, 3), cat_count) # randomly decides which category to be allocated
    # print(choosed_cat)
    for j in choosed_cat:
        cat = categories[j]
        data[cat][i] = 1
        
# data[data["category_b"] == 1]
data.head()
```
